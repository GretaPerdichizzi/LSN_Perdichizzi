```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize

def FileImport (nomefile=None, data_array=None, i=None):
    with open(nomefile) as fileinput:
        f=open(nomefile, 'r')
        lines=f.readlines()
        for x in lines:
            data_array.append(float(x.split()[i]))
        f.close()
```

In this exercise I resolved a Traveling Salesman Problem (TSP) using a genetic algorithm (GA).

Firstly I created the 2 initial configurations for the cities, which will remain unchanged during the simulation. I generated:

- 1 configuration (saved in "config.square") of 32 cities placed **inside** a square
- 1 configuration (saved in 2config.round") of 32 cities placed **on** a circumference

Below the configurations:

```python
x_round, y_round = [], []
x_square, y_square = [], []

FileImport("../Es9/config.square", x_square, 0)
FileImport("../Es9/config.square", y_square, 1)
FileImport("../Es9/config.round", x_round, 0)
FileImport("../Es9/config.round", y_round, 1)

fig=plt.figure(figsize=(14,14))
fig.suptitle('Cities configuration')
plt.subplot(2,2,1)
plt.plot(x_square,y_square, marker='o', markerfacecolor='r')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2,2,2)
plt.plot(x_round, y_round, marker='o', markerfacecolor='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

From the starting chromosome, by random permutation on the genes, I created the others. Each chromosome is a different permutation of the cities, in order to have the ordered path for the salesman.

I've selected *100* randomly shuffled chromosomes to be the starting population.

Than, by using the **cost/loss function**:
$$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$

I evaluated the total distance to be done by the salesman to complete each path. This distance is the **fitness** of the simulation we want to minimize. 

Than I **sorted** the chomosomes in Population class on the fitness basis: from which who had the shorter $L$ to the ones with the longer $L$.

For the **selection** operator I chose a **rigged-roulette** technique : I selected the individual $j$ with the algorithm: $j=N\times int(r^p)$ where $r$ is a uniform random number $r\in[0,1)$ and $p=2$. 

I wrote a **check function** to verify if the simulation bounds (no repetitions and no changes of the 1st city) are respected and I checked it on each creation of a new individual. 

Than I used the **crossover** operator and the **genetic-mutator** suggested in the given notebook with probability $p_{crossover}=80 \%$ and $p_{mutations}=15 \%$.

Below a picture of $L^{(1)}$ of the best path and of $\langle L^{(1)}\rangle$, averaged on the best half of the population, as a function of the iterations/generations of the algorithm, for both the world configurations:

```python
round_fitness_best, round_fitness_mean, round_ngen = [], [], []
square_fitness_best, square_fitness_mean, square_ngen = [], [], []

FileImport("../Es9/fitness_mean_round.out", round_ngen, 0)
FileImport("../Es9/fitness_mean_round.out", round_fitness_mean, 1)
FileImport("../Es9/fitness_mean_round.out", round_fitness_best, 2)
FileImport("../Es9/fitness_mean_square.out", square_ngen, 0)
FileImport("../Es9/fitness_mean_square.out", square_fitness_mean, 1)
FileImport("../Es9/fitness_mean_square.out", square_fitness_best, 2)


fig=plt.figure(figsize=(14,14))
fig.suptitle('Fitness functions')
plt.subplot(2,2,1)
plt.plot(round_ngen,round_fitness_mean, label="Mean")
plt.plot(round_ngen, round_fitness_best, label="Best")
plt.title("Round World")
plt.legend()
plt.xlabel('n generations')
plt.ylabel('fitness')


plt.subplot(2,2,2)
plt.plot(square_ngen,square_fitness_mean, label="Mean")
plt.plot(square_ngen, square_fitness_best, label="Best")
plt.title("Square World")
plt.legend()
plt.xlabel('n generations')
plt.ylabel('fitness')
plt.show()
```

Reported below also a picture of the best path for both the worlds cofigurations:

```python
x_round, y_round = [], []
x_square, y_square = [], []

FileImport("../Es9/config_round.final", x_round, 0)
FileImport("../Es9/config_round.final", y_round, 1)
FileImport("../Es9/config_square.final", x_square, 0)
FileImport("../Es9/config_square.final", y_square, 1)

fig=plt.figure(figsize=(14,14))
fig.suptitle('Best paths')
plt.subplot(2,2,1)
plt.plot(x_round,y_round, marker='o', markerfacecolor='r')
plt.scatter(x_round[0], y_round[0], s=150, marker='D', color='m', label="1st city")
plt.title("Round World")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,2)
plt.plot(x_square,y_square, marker='o', markerfacecolor='r')
plt.scatter(x_square[0], y_square[0], s=150, marker='D', color='m', label="1st city")
plt.title("Square World")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

As we can see in round world we reach the recognizabile global minimum, given from the sum of the circumference ropes. In the square world we aren't sure to have reached the global minimum, but probably this is a good minimum because we don't have intersections in the path.
