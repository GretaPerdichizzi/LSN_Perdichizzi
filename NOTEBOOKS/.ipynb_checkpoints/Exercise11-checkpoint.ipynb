```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

n_test=200

def LinearFit(sigma, n_train, n_epochs):
    m = 2
    b = 1
    
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, n_train)
    x_valid = np.random.uniform(-1, 1, n_test)
    x_valid.sort()
    y_target = m*x_valid +b
    
    y_train = np.random.normal(m * x_train +b, sigma)
    y_valid = np. random.normal(m * x_valid +b, sigma)
    
   # compose the NN model
    model = Sequential()
    model.add(Dense(1, input_shape=(1,)))

    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

    history = model.fit(x=x_train, y=y_train, 
              batch_size=32, epochs=n_epochs,
              shuffle=True,
              validation_data=(x_valid, y_valid), verbose=0)
    
    x_predicted = np.random.uniform(-1,1,100)
    y_predicted = model.predict(x_predicted)
    plt.figure(figsize=(10,5))
    plt.scatter(x_predicted, y_predicted, color ='r', label = 'Evaluated function')
    plt.plot(x_valid, y_target, label='Ideal function')
    plt.title('Linear fit')
    plt.legend()
    plt.grid(True)
    plt.show

    
def PolinomialFit(sigma, n_train, n_epochs, act_f, n_neurons, loss_f, optimizer, n_layers, inside):
    a=4
    b=-3
    c=-2
    d=3
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, n_train)
    x_valid = np.random.uniform(-1, 1, n_test)
    x_valid.sort()
    y_target = a + b*x_valid + c*(x_valid**2) + d*(x_valid**3)
    
    y_train = np.random.normal(a + b*x_train + c*(x_train**2) + d*(x_train**3))
    y_valid = np.random.normal(a + b*x_valid + c*(x_valid**2) + d*(x_valid**3))
    
    x_outside = np.random.uniform(-2, 2, n_test)
    model = Sequential()
    #input layer
    model.add(Dense(n_neurons, input_shape=(1,), activation=act_f))
    #hidden layers
    for i in range(1, n_layers-1):
        model.add(Dense(n_neurons, activation=act_f))
    #readout layer
    model.add(Dense(1))
    
    model.compile(optimizer=optimizer, loss=loss_f, metrics=['mse'])
    
    history = model.fit(x=x_train, y=y_train, 
              batch_size=32, epochs=n_epochs,
              shuffle=True, validation_data=(x_valid, y_valid), verbose=0)
    
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)
    
    x_predicted = np.random.uniform(-1,1,100)
    y_predicted = model.predict(x_predicted)
    y_predicted_outside = model.predict(x_outside)
    plt.figure(figsize=(10,5))
    if(inside==True):
        plt.scatter(x_predicted, y_predicted, color ='r', label = 'Evaluated function')
    else:
        plt.scatter(x_outside, y_predicted_outside, color ='r', label = 'Evaluated function')
        
    plt.plot(x_valid, y_target, label='Ideal function')
    plt.title('Polynomial fit with ' + str(n_layers) + ' layer(s) and ' + act_f + " as activation function")
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Neural Networks on linear regression

In this exercise I studied how does the linear regression discussed in the given Notebook 11 depend on the number of epochs, $N_{\mathrm{epochs}}$, the number of data points $N_{\mathrm{train}}$ and on the noise $\sigma$.  In order to do this I studied a set of data with different noise values, distributed around :
$$f(x)=2x+1$$
Below the model of the 1 neuron Network studied:

```python
m=2
b=1

np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, n_test)
x_valid.sort()


y_target = m * x_valid + b # ideal (target) linear function

y_train_mean = m * x_train + b # actual measures from which we want to guess regression parameters
y_valid_mean = m * x_valid + b

# compose the NN model
model = Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

I decided to study among different parameter values: 

- $N_{epochs} = [10, 50, 100]$
- $N_{train} = [100, 500, 1000]$
- $\sigma= [0.1, 0.3, 0.5]$

the configuration in which the loss function is the minimum. See the output of the code below to see the results:

```python
Ns_epochs = [10, 50, 100]
Ns_train = [100, 500, 1000]
sigmas = [0.1, 0.3, 0.5]
best_loss = 10000
best = [[0,0,0],[0,0,0], [0,0,0]]
i=0
for sigma in sigmas:
    for N_epochs in Ns_epochs:
        for N_train in Ns_train:
            y_train = np.random.normal(y_train_mean, sigma)
            y_valid = np.random.normal(y_valid_mean, sigma)
            
            history = model.fit(x=x_train[:N_train], y=y_train[:N_train], 
              batch_size=32, epochs=N_epochs,
              shuffle=True, # a good idea is to shuffle input before at each epoch
              validation_data=(x_valid, y_valid), verbose=0)
            
            if(history.history['val_loss'][-1]<best_loss):
                best_loss=history.history['val_loss'][-1]
                best_N_epochs, best_N_train = N_epochs, N_train
                
    print("Best validation-loss found for sigma = ", sigma, 
          ": N_train =", best_N_train, ", N_epochs =", best_N_epochs)
    best_loss = 10000
    best[i][0]=sigma
    best[i][1]=best_N_train
    best[i][2]=best_N_epochs
    i=i+1
```

Below the pictures of the estimated functions with these values:

```python
print ('sigma = '+str(best[0][0])+' N_train = '+str(best[0][1])+' N_epochs= '+str(best[0][2]))
LinearFit(best[0][0],best[0][1],best[0][2])
```

```python
print ('sigma = '+str(best[1][0])+' N_train = '+str(best[1][1])+' N_epochs = '+str(best[1][2]))
LinearFit(best[1][0],best[1][1],best[1][2])
```

```python
print ('sigma = '+str(best[2][0])+' N_train = '+str(best[2][1])+' N_epochs = '+str(best[2][2]))
LinearFit(best[2][0],best[2][1],best[2][2])
```

As we can see if the noise is low we can use less training points and less training epochs. 

If we have few data for the training I must increase the number of training epochs. Sometimes however it could mean that we are training for a long period on a small set of training data and there is the risk of *overfitting* : obtaing better results for data which belongs to the training set but worste on the new one.

# <span style="color:red"> Exercise 11.3 </span>

```python
For this exercise I extended the model to fit the simple trigonometric 2D function 

$$f(x,y) = \sin(x^2+y^2)$$ 

in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

On the base of the previous studies I decided to choose the following parameters for the NN:

- Number of trials points = 13000
- Number of epochs = 100
- Noise = 0.2
- Number of neurons in each layer = 30
- Activation function = relu (with the tanh the loss function was oscillating too much)
- Loss function = mean square error
- Opitmizer = sgd
- Number of layers = 5
```

```python
x_train = np.random.uniform(-1.5,1.5,13000)
y_train = np.random.uniform(-1.5,1.5,13000)
x_valid = np.random.uniform(-1.5,1.5,1000)
y_valid = np.random.uniform(-1.5,1.5,1000)

z_target = np.sin(x_valid**2+y_valid**2)

sigma = 0.2
z_train = np.random.normal(np.sin(x_train**2+y_train**2), sigma)
z_valid = np.random.normal(np.sin(x_valid**2+y_valid**2), sigma)

xy_train = np.array([x_train, y_train]).T
xy_valid = np.array([x_valid, y_valid]).T

model = Sequential()
#input layer
model.add(Dense(30, input_shape=(2,), activation='relu'))
#hidden layers
for i in range(1, 4):
    model.add(Dense(30, activation='relu'))
#readout layer
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.summary()

history = model.fit(x=xy_train, y=z_train, 
          batch_size=32, epochs=100,
          shuffle=True, validation_data=(xy_valid, z_valid), verbose=0)
    
plt.figure(figsize = (14,7))
plt.plot(history.history['loss'], label='Train-loss')
plt.plot(history.history['val_loss'], label='val-loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```

```python
fig = plt.figure(figsize = (14,10))
ax = fig.gca(projection='3d')

ax.plot_trisurf(x_valid, y_valid, z_target, linewidth=0.2, antialiased=True, cmap=cm.PuBu)
x_predicted = np.random.uniform(-1.5,1.5,500)
y_predicted = np.random.uniform(-1.5,1.5,500)
xy_predicted = np.array([x_predicted, y_predicted]).T
z_predicted = model.predict(xy_predicted)
ax.scatter(x_predicted, y_predicted, z_predicted, color = 'r', s=3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
