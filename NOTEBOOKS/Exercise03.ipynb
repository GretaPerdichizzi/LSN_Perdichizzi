```python
import csv
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def FileImport (nomefile=None, data_array=None):
    with open(nomefile) as fileinput:
        lettore = csv.reader(fileinput,delimiter="\t", quoting=csv.QUOTE_NONNUMERIC)
        for riga in lettore:
            for dato in riga:
                if type(dato)==float:
                    dato = dato
                    data_array.append(dato)
                    
def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)
```

# Computing Plain vanilla options by MC simulations

The sampling of an asset price $S(t)$ is the same of a $GBM(r,\sigma^2)$. So I simulate $M=10^4$ times $S(T)$ directly sampled with the equation:

$$S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$

and $M=10^4$ times $S(T)$ sampled with a discretized $GBM(r,\sigma^2)$ with the following recursive way:

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$

The pay-off function at time $T$ is $$(S(T)-K)^+:=max[0, S(T)-K]$$
I have to discount the profit of a factor $e^{-rT}$, due to the hypothetical interest that a bank would have guaranteed with a deposit at time $t_0=0$. So to evaluate an European option price I will check the difference $S(T)-K$. 
If it is positive $S(T)-K>0$, the max corresponds to $S(T)-K$ and I have a call-option price 

$$C[S(0),0]=e^{-rT}(S(T)-K)$$ 

If it is negative $S(T)-K<0$ the max is $K-S(T)$ and I have a put-option price 

$$P[S(0),0]=e^{-rT}(K-S(T))$$

Below I import the values of $C[S(0),0]$ and $P[S(0),0]$ simulated like above with a c++ code and I estimate the mean and the uncertainty with the blocks method, dividing $M$ in $100$ blocks $N=100$.

I use the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

```python
M=1000000              # Total number of throws
N=100                  # Number of blocks
L=int(M/N)             # Number of throws in each block
x = np.arange(N)       # [0,1,2,...,N-1]

#Importing data from direct simulation 
C_dir, P_dir = [], [] 
C_discr, P_discr = [], []
FileImport("../Es3/c_dir", C_dir)
FileImport("../Es3/p_dir", P_dir)
FileImport("../Es3/c_discr", C_discr)
FileImport("../Es3/p_discr", P_discr)


ave_C_dir, av2_C_dir, ave_P_dir, av2_P_dir = [], [], [], []
ave_C_discr, av2_C_discr, ave_P_discr, av2_P_discr = [], [], [], []
sum_C_dir_prog = np.zeros(N)
su2_C_dir_prog = np.zeros(N)
err_C_dir_prog = np.zeros(N)
sum_P_dir_prog = np.zeros(N)
su2_P_dir_prog = np.zeros(N)
err_P_dir_prog = np.zeros(N)
sum_C_discr_prog = np.zeros(N)
su2_C_discr_prog = np.zeros(N)
err_C_discr_prog = np.zeros(N)
sum_P_discr_prog = np.zeros(N)
su2_P_discr_prog = np.zeros(N)
err_P_discr_prog = np.zeros(N)

for i in range(N):
    sum_C_dir_temp=0
    sum_P_dir_temp=0
    sum_C_discr_temp=0
    sum_P_discr_temp=0
    for j in range(L):
        k = j+i*L
        sum_C_dir_temp+=C_dir[k]
        sum_P_dir_temp+=P_dir[k]
        sum_C_discr_temp+=C_discr[k]
        sum_P_discr_temp+=P_discr[k]
    ave_C_dir.append(sum_C_dir_temp/L)
    av2_C_dir.append(ave_C_dir[i]**2)
    ave_P_dir.append(sum_P_dir_temp/L)
    av2_P_dir.append(ave_P_dir[i]**2)
    ave_C_discr.append(sum_C_discr_temp/L)
    av2_C_discr.append(ave_C_discr[i]**2)
    ave_P_discr.append(sum_P_discr_temp/L)
    av2_P_discr.append(ave_P_discr[i]**2)
    
for i in range(N):
    for j in range(i+1):
        sum_C_dir_prog[i] += ave_C_dir[j] 
        su2_C_dir_prog[i] += av2_C_dir[j] 
        sum_P_dir_prog[i] += ave_P_dir[j] 
        su2_P_dir_prog[i] += av2_P_dir[j]
        sum_C_discr_prog[i] += ave_C_discr[j] 
        su2_C_discr_prog[i] += av2_C_discr[j]
        sum_P_discr_prog[i] += ave_P_discr[j] 
        su2_P_discr_prog[i] += av2_P_discr[j]
        
    sum_C_dir_prog[i]/=(i+1) 
    su2_C_dir_prog[i]/=(i+1) 
    err_C_dir_prog[i] = error(sum_C_dir_prog,su2_C_dir_prog,i) 
    sum_P_dir_prog[i]/=(i+1) 
    su2_P_dir_prog[i]/=(i+1) 
    err_P_dir_prog[i] = error(sum_P_dir_prog,su2_P_dir_prog,i) 
    sum_C_discr_prog[i]/=(i+1) 
    su2_C_discr_prog[i]/=(i+1) 
    err_C_discr_prog[i] = error(sum_C_discr_prog,su2_C_discr_prog,i) 
    sum_P_discr_prog[i]/=(i+1) 
    su2_P_discr_prog[i]/=(i+1) 
    err_P_discr_prog[i] = error(sum_P_discr_prog,su2_P_discr_prog,i) 

picture1=plt.figure(figsize=(15, 10))
a1=plt.subplot(2, 2, 1)
plt.errorbar(x,sum_C_dir_prog,yerr=err_C_dir_prog, label='MC')
plt.axhline(y=14.976, label='Black-Scholes', color='r', linestyle='--', linewidth=1)
plt.title('Direct European call-option')
plt.legend()
plt.xlabel('#Blocks')
plt.ylabel('Price')
plt.grid(True)

a2=plt.subplot(2, 2, 2)
plt.errorbar(x,sum_P_dir_prog,yerr=err_P_dir_prog, label='MC')
plt.axhline(y=5.460, label='Black-Scholes', color='r', linestyle='--', linewidth=1)
plt.title('Direct European put-option')
plt.legend()
plt.xlabel('#Blocks')
plt.ylabel('Price')
plt.grid(True)

a3=plt.subplot(2, 2, 3)
plt.errorbar(x,sum_C_discr_prog,yerr=err_C_discr_prog, label='MC')
plt.axhline(y=14.976, label='Black-Scholes', color='r', linestyle='--', linewidth=1)
plt.title('Discretized European call-option')
plt.legend()
plt.xlabel('#Blocks')
plt.ylabel('Price')
plt.grid(True)

a4=plt.subplot(2, 2, 4)
plt.errorbar(x,sum_P_discr_prog,yerr=err_P_discr_prog, label='MC')
plt.axhline(y=5.460, label='Black-Scholes', color='r', linestyle='--', linewidth=1)
plt.title('Discretized put-option')
plt.legend()
plt.xlabel('#Blocks')
plt.ylabel('Price')
plt.grid(True)

plt.show()

final = [0,0,0,0]
final[0]= round(sum_C_dir_prog[N-1], 2)
final[1]= round(sum_P_dir_prog[N-1], 2)
final[2]= round(sum_C_discr_prog[N-1], 2)
final[3]= round(sum_P_discr_prog[N-1], 2)

error = [0,0,0,0]
error[0]= round(err_C_dir_prog[N-1], 2)
error[1]= round(err_P_dir_prog[N-1], 2)
error[2]= round(err_C_discr_prog[N-1], 2)
error[3]= round(err_P_discr_prog[N-1], 2)

strings = ["Call direct = ", "Put direct = ", "Call discr = ", "Put discr = "]

for i in range(4):
    print(strings[i]+str(final[i])+" +/- "+str(error[i]))
```

We can see that the final values for a Call-option and a Put-option are acceptable according to the exact analytic results of the Black&Scholes solution, both in direct case and in discretized case.
