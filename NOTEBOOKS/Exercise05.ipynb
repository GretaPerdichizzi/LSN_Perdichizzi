```python
import csv
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def FileImport (nomefile=None, data_array=None, i=None):
    with open(nomefile) as fileinput:
        f=open(nomefile, 'r')
        lines=f.readlines()
        for x in lines:
            data_array.append(float(x.split('\t')[i]))
        f.close()
                    
def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)
```

# Metropolis algorithm on hydrogen atom simulations

Fot this exercise I used the Metropolis algorithm to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ **in Cartesian coordinates** using an uniform transition probability $T(\vec{x}|\vec{y})$. 

I also used the sampled positions to estimate $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$. I used as usual data blocking and I gave an estimate of the statistical uncertainties.

Below the trasformations of  $|\Psi_{1,0,0}(r,\theta,\phi)|^2$ and  $|\Psi_{2,1,0}(r,\theta,\phi)|^2$ in cartesian coordinates:
$$\Psi_{1,0,0}(x,y,z)= \frac{a_0^{-\frac{3}{2}}}{\sqrt{\pi}}e^{-\frac{1}{a_0}\sqrt{x^2+y^2+z^2}}$$
and
$$\Psi_{2,1,0}(x,y,z)= \frac{a_0^{-\frac{5}{2}}}{8}\sqrt{\frac{2}{\pi}}ze^{-\frac{1}{2a_0}\sqrt{x^2+y^2+z^2}}$$

I choose the starting points of the simulations so that they are at a point where $p(x)=|\Psi_{n,l,m}(x,y,z)|^2$ is maximum, i.e. $x_n=(0, 0, 0)$ for $|\Psi_{1,0,0}(x,y,z)|^2$ and $x_n=(1.73, 0, 1)$ for $|\Psi_{2,1,0}(x,y,z)|^2$. In this way we can have a better efficiency for the algorithm. 

If I had chosen starting points far from high $p(x)$ zones, the simulations would have needed moore equilibration time.

For the Metropolis algorithm we have to firstly generate a (set of) coodinate(s) $x'$, I choose to change all 3 coordinates in 1 Metropolis evaluation, so I have to generate $x',y',z'$. For the lenght of the interval of uniform number generation we can follow the **50% empiric rule**.

Since the transition probability required $T(x|y)$ is uniform, and therefore it's symmetric, we can evaluate the acceptance probability $A(x'|x_n)$ with

$$A(x'|x_n)=min\left \{ 1, \frac{p(x')}{p(x_n)} \right \}$$


```python
M=1000000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)             # Number of throws in each block
x = np.arange(N)       # [0,1,2,...,N-1]

r100 = []
r210 = []
ave = np.zeros(N)
av2 = np.zeros(N)
su2_prog_100 = np.zeros(N)
su2_prog_210 = np.zeros(N)
sum_prog_100 = np.zeros(N)
err_prog_100 = np.zeros(N)
sum_prog_210 = np.zeros(N)
err_prog_210 = np.zeros(N)

FileImport ('../Es5/psi100_uniform.dat', r100, 3)
FileImport ('../Es5/psi210_uniform.dat', r210, 3)

for i in range(N):
    sum = 0
    for j in range(L):
        k = j+i*L
        sum += r100[k]
    ave[i] = sum/L       # r_i 
    av2[i] = (ave[i])**2 # (r_i)^2 
for i in range(N):
    for j in range(i+1):
        sum_prog_100[i] += ave[j] # SUM_{j=0,i} r_j
        su2_prog_100[i] += av2[j] # SUM_{j=0,i} (r_j)^2
    sum_prog_100[i]/=(i+1) # Cumulative average
    su2_prog_100[i]/=(i+1) # Cumulative square average
    err_prog_100[i] = error(sum_prog_100,su2_prog_100,i) # Statistical uncertainty

#210
for i in range(N):
    sum = 0
    for j in range(L):
        k = j+i*L
        sum += r210[k]
    ave[i] = sum/L       # r_i 
    av2[i] = (ave[i])**2 # (r_i)^2 
i=0
j=0
for i in range(N):
    for j in range(i+1):
        sum_prog_210[i] += ave[j] # SUM_{j=0,i} r_j
        su2_prog_210[i] += av2[j] # SUM_{j=0,i} (r_j)^2
    sum_prog_210[i]/=(i+1) # Cumulative average
    su2_prog_210[i]/=(i+1) # Cumulative square average
    err_prog_210[i] = error(sum_prog_210,su2_prog_210,i) # Statistical uncertainty
    
#plots
picture1=plt.figure(figsize=(15, 10))
a1=plt.subplot(2, 2, 1)
plt.errorbar(x,sum_prog_100,yerr=err_prog_100, label='Estimated value')
plt.axhline(1.5, color='r', linestyle='--', label='Expected value')
plt.title('Ground wave function')
plt.legend()
plt.xlabel('#blocks')
plt.ylabel('<r>')
plt.grid(True)
a2=plt.subplot(2, 2, 2)
plt.errorbar(x,sum_prog_210,yerr=err_prog_210, label='Estimated value')
plt.axhline(5, color='r', linestyle='--', label='Expected value')
plt.title('2p wave function')
plt.legend()
plt.xlabel('#blocks')
plt.ylabel('<r>')
plt.grid(True)
plt.show()

#print values
print ("UNIFORM GENERATION")
print ("Mean of the radius for the ground wave function = " +str(round(sum_prog_100[N-1],3))+" +/- "+str(round(err_prog_100[N-1],3)))
print ("Mean of the radius for the 2p wave function = " +str(round(sum_prog_210[N-1],2))+" +/- "+str(round(err_prog_210[N-1],2)))
```
